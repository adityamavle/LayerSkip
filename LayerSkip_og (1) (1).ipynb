```python
import torch
import time
import pandas as pd
from transformers import AutoModelForCausalLM, AutoTokenizer, GenerationConfig
from fvcore.nn import FlopCountAnalysis
from IPython.display import display
```

```python
# 1) Compute FLOPs per token
def compute_flops_per_token(model, seq_len=1):
    model.eval()
    # ── Disable the past-key-values cache so the tracer only sees tensors ──
    model.config.use_cache = False

    dummy = torch.ones(1, seq_len, dtype=torch.long).cuda()
    # pass inputs as a tuple
    flops = FlopCountAnalysis(model, (dummy,))
    return flops.total() / seq_len
```

```python
def measure_time_per_token(model, input_ids, max_new_tokens=50, warmup=3):
    model.eval()
    # warm-up
    for _ in range(warmup):
        model.generate(input_ids, max_new_tokens=max_new_tokens)
    torch.cuda.synchronize()
    start = torch.cuda.Event(enable_timing=True)
    end   = torch.cuda.Event(enable_timing=True)
    start.record()
    out = model.generate(input_ids, max_new_tokens=max_new_tokens, output_hidden_states=False)
    end.record()
    torch.cuda.synchronize()
    elapsed_ms = start.elapsed_time(end)
    return elapsed_ms / max_new_tokens, out
```

```python
# 3) Strategies to benchmark
strategies = {
    'base': {
        'model_id': 'meta-llama/Llama-3.2-1B',
        'gen_kwargs': {}
    },
    'layerdrop': {
        'model_id': 'facebook/layerskip-llama3.2-1B',
        'gen_kwargs': {
            'generation_strategy':'layerdrop',
            'layerdrop_rate':0.5
        }
    },
    'self_speculative': {
        'model_id': 'facebook/layerskip-llama3.2-1B',
        'gen_kwargs': {
            'generation_strategy':'self_speculative',
            'exit_layer':8,
            'num_speculations':6
        }
    },
    'depth_adaptive': {
        'model_id': 'facebook/layerskip-llama3.2-1B',
        'gen_kwargs': {
            'generation_strategy': 'depth_adaptive',
            'threshold': 0.5,        # ← stop once confidence gap ≥ 0.5
        }},
}
```

```python
# 4) Prepare tokenizer & prompt
tokenizer = AutoTokenizer.from_pretrained('facebook/layerskip-llama3.2-1B')
prompt = "The meaning of life is"
input_ids = tokenizer(prompt, return_tensors='pt').input_ids.cuda()
```

```python
# 5) Run benchmarks
records = []
for name, cfg in strategies.items():
    print(f"\n▶ Profiling '{name}'…")
    model = AutoModelForCausalLM.from_pretrained(cfg['model_id']).cuda()
    model.generation_config = GenerationConfig(**cfg['gen_kwargs'])

    flops_per_token = compute_flops_per_token(model)
    ms_per_token, output = measure_time_per_token(model, input_ids, max_new_tokens=50)

    # capture LayerSkip’s exit layers if provided
    exit_layers = getattr(output, 'exit_layers', None)
    avg_exit = (sum(exit_layers)/len(exit_layers)) if exit_layers is not None else model.config.num_hidden_layers

    records.append({
        'Strategy':        name,
        'FLOPs_per_token': float(flops_per_token * (avg_exit if name!='base' else model.config.num_hidden_layers)),
        'Avg_exit_layer':  avg_exit,
        'ms_per_token':    ms_per_token
    })
```

```python
df = pd.DataFrame(records)

# 7) Display in notebook (rich HTML)
display(df)
```
